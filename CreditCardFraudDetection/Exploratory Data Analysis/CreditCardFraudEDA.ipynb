# Welcome to the Credit Card Fraud Detection project! 

The dataset provided focuses on credit card transactions in September 2013 by European cardholders. Its primary objective is to aid credit card companies in detecting fraudulent transactions, ensuring customers are not wrongly charged for unauthorized purchases.

The dataset encompasses transactions spanning two days, encompassing a total of 284,807 transactions. Out of these, 492 are identified as fraudulent, constituting only 0.172% of the total. 

The dataset exclusively consists of numerical input variables derived from Principal Component Analysis (PCA) transformation. Due to confidentiality constraints, the original features and additional background information cannot be disclosed. The features labeled as V1 through V28 represent the confidential features, with 'Time' and 'Amount' being the only unaltered features. 'Time' signifies the seconds elapsed since the first transaction, while 'Amount' indicates the transaction amount. Finally, the 'Class' feature serves as the response variable, assuming a value of 1 in cases of fraud and 0 otherwise.

```python
# Checking if the dataset is balanced
distribution = df['Class'].value_counts()

# Visualizing the class balance
distribution.plot(kind='bar')
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Distribution of Class After Undersampling')
plt.show()


class_counts = df['Class'].value_counts()
print(class_counts) # the dataset in very unbalanced
```

```python
# The dataset is unbalanced, the class must be balanced. for that we will preform random undersampling
from imblearn.under_sampling import RandomUnderSampler

X = df.drop(columns=['Class'])
y = df['Class']

sampler = RandomUnderSampler(random_state=42)
X_resampled, y_resampled = sampler.fit_resample(X, y)

# Saving the balanced dataframe
df_balanced = pd.concat([X_resampled, y_resampled], axis=1)
df_balanced.to_csv('C:/Users/yomna/codsoft/CreditCardFraudDetection/Dataset/card_balanced_data.csv', index=False)
```

```python
# Visualizing the class balance after Undersampling
plt.bar(y_resampled.unique(), y_resampled.value_counts(), color=['plum', 'lightblue'])
plt.xticks(y_resampled.unique())
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Distribution of Class After Undersampling')
plt.show()
```

```python
# Checking the connection between Time and Class
df[df['Class'] == 0]['Time'].plot(kind='kde', label='Class 0')
df[df['Class'] == 1]['Time'].plot(kind='kde', label='Class 1')
plt.legend(loc='upper right')
plt.title('Density Plot of Time by Class')
plt.xlabel('Time')
plt.ylabel('Density')
plt.show()
```

```python
# Checking the connection between Amount and Class
plt.figure(figsize=(10, 6))

plt.scatter(df[df['Class'] == 0]['Time'], df[df['Class'] == 0]['Amount'], color='blue', label='Class 0', alpha=0.5)
plt.scatter(df[df['Class'] == 1]['Time'], df[df['Class'] == 1]['Amount'], color='red', label='Class 1', alpha=0.5)

plt.title('Scatter Plot of Time vs Amount by Class')
plt.xlabel('Time (seconds)')
plt.ylabel('Amount')
plt.legend()
plt.show()
```
